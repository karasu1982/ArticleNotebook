```python
# データ取得と前処理

from completejourney_py import get_data
cj = get_data()
transactions = cj['transactions']
products     = cj['products']
demographics = cj['demographics']
```

```python
transactions.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.cluster import KMeans
import numpy as np

# 結合
df = transactions.merge(products, on='product_id', how='left')\
                 .merge(demographics, on='household_id', how='left')

# 単価列と基本クレンジング
df['unit_price'] = (df['sales_value'] / df['quantity']).replace([np.inf, -np.inf], np.nan)
df = df[(df['quantity'] > 0) & (df['sales_value'] > 0)]
df = df.dropna(subset=['unit_price'])

# household_id×weekで週次集計
agg_keys = ['household_id', 'week']
weekly_num = df.groupby(agg_keys).agg(
    basket_size = ('quantity', 'sum'),
    total_sales = ('sales_value', 'sum'),
    mean_price  = ('unit_price', 'mean')
).reset_index()

# カテゴリ属性の最頻値
def mode_or_nan(x):
    return x.mode().iloc[0] if not x.mode().empty else np.nan

demo_cols = ['age', 'marital_status', 'household_size', 'household_comp', 'kids_count']
weekly_cat = df.groupby(agg_keys)[demo_cols].agg(mode_or_nan).reset_index()

weekly = weekly_num.merge(weekly_cat, on=agg_keys, how='left')

# 顧客クラスタ（属性OneHot→KMeans）
cat_for_cluster = demo_cols
enc = OneHotEncoder(handle_unknown='ignore')
Xc  = enc.fit_transform(weekly[cat_for_cluster].astype(str).fillna('NA'))

n_clusters = 100  # Colab向けに軽め。論文例では~400
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
weekly['customer_cluster'] = kmeans.fit_predict(Xc)

# 学習用テーブル
final_cols = ['household_id', 'week', 'basket_size', 'total_sales', 'mean_price', 'customer_cluster'] + cat_for_cluster
data = weekly[final_cols].dropna().set_index(['household_id', 'week']) # Set index and keep it


print("Final dataset shape:", data.shape)
data.head()
```

```python
# データ分割

from sklearn.model_selection import train_test_split

train_temp, E = train_test_split(data, test_size=0.2, random_state=42)   # Eval
T, H = train_test_split(train_temp, test_size=0.25, random_state=42)     # T:60% / H:20%

print("Train:", T.shape, "Holdout:", H.shape, "Eval:", E.shape)
```

```python
cat_cols = ['customer_cluster', 'age', 'marital_status', 'household_size', 'household_comp', 'kids_count']
num_cols = ['basket_size', 'total_sales', 'mean_price']
```

```python
# 合成データ生成（CTGAN）

import pandas as pd

from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata

# 例：前段で作った学習用データ
# T: 学習用DataFrame
# num_cols, cat_cols, final_cols: 既出の列リストを流用

# 1) pandasのnullable型を通常型に揃える（学習の安定化）
T_clean = T.copy()
for c in T_clean.columns:
    if pd.api.types.is_integer_dtype(T_clean[c]):
        T_clean[c] = T_clean[c].astype('int64')
    elif pd.api.types.is_float_dtype(T_clean[c]):
        T_clean[c] = T_clean[c].astype('float64')
    else:
        T_clean[c] = T_clean[c].astype('object')

# 2) メタデータを生成・自動推定 → カラム型を上書き
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=T_clean)

# 数値/カテゴリの型を明示（特にカテゴリはcategoricalに）
for col in num_cols:
    if col in T_clean.columns:
        metadata.update_column(col, sdtype='numerical')
for col in cat_cols:
    if col in T_clean.columns:
        metadata.update_column(col, sdtype='categorical')

# 主キーがあれば指定（なければスキップ）
# 例: household_id を持っているなら
# metadata.set_primary_key('household_id')

# 3) CTGANで学習・生成
synth = CTGANSynthesizer(metadata, epochs=100)
synth.fit(T_clean)
S = synth.sample(num_rows=len(T_clean))

# 4) 列順を合わせる（評価コードとの整合のため）
S = S[[c for c in final_cols if c in S.columns]]

S.head()
```

## 忠実性：周辺分布（Wasserstein/JS）＋結合（相関/依存）の距離

周辺分布（数値=WD、カテゴリ=JS）と結合（数値×数値=相関、カテゴリ×カテゴリ=Theil’s U、数値×カテゴリ=相関比）の距離で忠実性を定量化。Hに近いほど理想。

```python
from itertools import combinations
from scipy.spatial.distance import jensenshannon
from scipy.stats import wasserstein_distance, entropy

def jensen_shannon_distance(p, q):
    p = np.asarray(p, dtype=float); p = p / (p.sum() + 1e-12)
    q = np.asarray(q, dtype=float); q = q / (q.sum() + 1e-12)
    return jensenshannon(p, q, base=2)

def theils_u(x, y):
    s_xy = pd.crosstab(x, y) + 1e-12
    p_xy = s_xy / s_xy.sum().sum()
    p_x  = p_xy.sum(axis=1)
    p_y  = p_xy.sum(axis=0)
    H_y  = entropy(p_y, base=2)
    H_y_given_x = 0.0
    for xi in p_x.index:
        p_y_given_x = p_xy.loc[xi, :] / p_x.loc[xi]
        H_y_given_x += p_x.loc[xi] * entropy(p_y_given_x, base=2)
    return (H_y - H_y_given_x) / (H_y + 1e-12)

def correlation_ratio(categories, measurements):
    cats = pd.Categorical(categories).codes
    y = np.asarray(measurements, dtype=float)
    total_mean = np.mean(y)
    ss_between = 0.0
    for c in np.unique(cats):
        y_c = y[cats == c]
        if len(y_c) == 0:
            continue
        ss_between += len(y_c) * (np.mean(y_c) - total_mean)**2
    ss_total = np.sum((y - total_mean)**2) + 1e-12
    return ss_between / ss_total

def marginal_distances(A, B):
    wd = [wasserstein_distance(A[col].values, B[col].values) for col in num_cols]
    js = []
    for col in cat_cols:
        p = A[col].value_counts().sort_index()
        q = B[col].value_counts().sort_index()
        idx = sorted(set(p.index).union(set(q.index)))
        p = p.reindex(idx, fill_value=0).values
        q = q.reindex(idx, fill_value=0).values
        js.append(jensen_shannon_distance(p, q))
    return np.mean(wd), np.mean(js)

def joint_distances(A, B):
    # 数値-数値（Pearson相関行列のL2距離）
    corr_A = A[num_cols].corr().values.flatten()
    corr_B = B[num_cols].corr().values.flatten()
    num_num = np.linalg.norm(corr_A - corr_B)

    # カテゴリ-カテゴリ（Theil’s U の対称平均の行列→L2）
    C = cat_cols
    M_A, M_B = [], []
    for i in range(len(C)):
        row_A, row_B = [], []
        for j in range(len(C)):
            xA, yA = A[C[i]], A[C[j]]
            xB, yB = B[C[i]], B[C[j]]
            uA = 0.5*(theils_u(xA, yA)+theils_u(yA, xA))
            uB = 0.5*(theils_u(xB, yB)+theils_u(yB, xB))
            row_A.append(uA); row_B.append(uB)
        M_A.append(row_A); M_B.append(row_B)
    M_A = np.array(M_A).flatten(); M_B = np.array(M_B).flatten()
    cat_cat = np.linalg.norm(M_A - M_B)

    # 数値-カテゴリ（相関比の行列→L2）
    NC_A = [correlation_ratio(A[c], A[n]) for c in cat_cols for n in num_cols]
    NC_B = [correlation_ratio(B[c], B[n]) for c in cat_cols for n in num_cols]
    num_cat = np.linalg.norm(np.array(NC_A) - np.array(NC_B))

    return num_num, cat_cat, num_cat

# T vs H, T vs S を比較
wd_H, js_H = marginal_distances(T, H)
wd_S, js_S = marginal_distances(T, S)
nnum_H, ccat_H, ncat_H = joint_distances(T, H)
nnum_S, ccat_S, ncat_S = joint_distances(T, S)

fidelity_df = pd.DataFrame({
    'dataset': ['Holdout(H)', 'Synthetic(S)'],
    'marginal_numeric_wasserstein': [wd_H, wd_S],
    'marginal_categorical_js': [js_H, js_S],
    'joint_num_num_l2': [nnum_H, nnum_S],
    'joint_cat_cat_l2': [ccat_H, ccat_S],
    'joint_num_cat_l2': [ncat_H, ncat_S],
})

fidelity_df
```

## 有用性：分類タスク（バスケットサイズ>10）

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_score, recall_score
from sklearn.preprocessing import OneHotEncoder
import numpy as np

target = (E['basket_size'] > 10).astype(int)

# Eの特徴（数値+カテゴリOneHot）
X_num = E[num_cols].copy()
X_cat = E[cat_cols].astype(str).fillna('NA')
enc_util = OneHotEncoder(handle_unknown='ignore')
X_cat_oh = enc_util.fit_transform(X_cat).toarray() # Convert sparse matrix to dense array
X_eval = np.hstack([X_num.values, X_cat_oh])

def fit_and_eval(train_df, name="Train(T)"):
    y = (train_df['basket_size'] > 10).astype(int)
    Xn = train_df[num_cols]
    Xc = train_df[cat_cols].astype(str).fillna('NA')
    oh = enc_util.transform(Xc).toarray() # Convert sparse matrix to dense array
    X = np.hstack([Xn.values, oh])
    clf = BaggingClassifier(n_estimators=100, random_state=42)
    clf.fit(X, y)
    pred = clf.predict(X_eval)
    proba = clf.predict_proba(X_eval)[:,1]
    return {
        'source': name,
        'accuracy': accuracy_score(target, pred),
        'f1': f1_score(target, pred),
        'roc_auc': roc_auc_score(target, proba),
        'precision': precision_score(target, pred),
        'recall': recall_score(target, pred)
    }

u_T = fit_and_eval(T, "Train(T)")
u_H = fit_and_eval(H, "Holdout(H)")
u_S = fit_and_eval(S, "Synthetic(S)")

utility_df = pd.DataFrame([u_T, u_H, u_S])
utility_df
```

## 有用性：商品アソシエーション（Apriori, Lift/Conviction）

```python
# household_id×weekで department の購入有無をブール化
def make_baskets(original_df):
    grp = original_df.groupby(['household_id','week','department'])['quantity'].sum().reset_index()
    grp['bought'] = (grp['quantity'] > 0).astype(int)
    basket = grp.pivot_table(index=['household_id','week'], columns='department', values='bought', fill_value=0)
    basket = basket.astype(bool)
    return basket

from mlxtend.frequent_patterns import apriori, association_rules
import numpy as np

# Create baskets from the original dataframe
full_basket = make_baskets(df)

# Filter baskets based on household_id and week in T, H, and S
# basket_T = full_basket.loc[T.set_index(['household_id', 'week']).index]
# basket_H = full_basket.loc[H.set_index(['household_id', 'week']).index]
# basket_S = full_basket.loc[S.set_index(['household_id', 'week']).index]
basket_T = full_basket.loc[T.index]
basket_H = full_basket.loc[H.index]
basket_S = full_basket.loc[S.index]

def apriori_rules(basket_bool, min_support=0.02):
    frequent = apriori(basket_bool, min_support=min_support, use_colnames=True)
    if frequent.empty:
        return pd.DataFrame()
    rules = association_rules(frequent, metric="lift", min_threshold=1.0)
    if rules.empty:
        return pd.DataFrame()
    return rules[['antecedents','consequents','support','confidence','lift','conviction']].sort_values('lift', ascending=False)

rules_T = apriori_rules(basket_T)
rules_H = apriori_rules(basket_H)
rules_S = apriori_rules(basket_S)

def summarize_rules(rules_df, label):
    if rules_df is None or rules_df.empty:
        return pd.Series({'label': label, 'n_rules': 0, 'mean_lift': np.nan, 'mean_conviction': np.nan})
    return pd.Series({
        'label': label,
        'n_rules': len(rules_df),
        'mean_lift': rules_df['lift'].mean(),
        'mean_conviction': rules_df['conviction'].replace(np.inf, np.nan).mean()
    })

assoc_summary = pd.DataFrame([
    summarize_rules(rules_T, 'Train(T)'),
    summarize_rules(rules_H, 'Holdout(H)'),
    summarize_rules(rules_S, 'Synthetic(S)')
])
assoc_summary
```

## プライバシー：DCR（最近傍距離の相対値）＆ CCR

```python
def encode_for_distance(df_in):
    num = df_in[num_cols].copy()
    cat = df_in[cat_cols].astype(str).fillna('NA')
    scaler = MinMaxScaler()
    num_scaled = scaler.fit_transform(num)
    ohe = OneHotEncoder(handle_unknown='ignore')
    cat_oh = ohe.fit_transform(cat).toarray()
    X = np.hstack([num_scaled, cat_oh])
    return X, scaler, ohe

def transform_with_enc(df_in, scaler, ohe):
    num = df_in[num_cols].copy()
    cat = df_in[cat_cols].astype(str).fillna('NA')
    X = np.hstack([scaler.transform(num), ohe.transform(cat).toarray()])
    return X

X_T, scaler_T, ohe_T = encode_for_distance(T)
X_H = transform_with_enc(H, scaler_T, ohe_T)
X_S = transform_with_enc(S, scaler_T, ohe_T)

def avg_min_l1(A, B, sample_B=5000):
    if len(B) > sample_B:
        idx = np.random.choice(len(B), size=sample_B, replace=False)
        B = B[idx]
    dmins = []
    for a in A:
        d = np.abs(B - a).sum(axis=1)
        dmins.append(d.min())
    return float(np.mean(dmins))

# DCR相対化：H→T を 1.0 基準に S→T を相対比較
dcr_HT = avg_min_l1(X_H, X_T)
dcr_ST = avg_min_l1(X_S, X_T)
rel_DCR_S = dcr_ST / (dcr_HT + 1e-12)
print("DCR(H->T) baseline = {:.3f} (set to 1.0)".format(dcr_HT))
print("DCR(S->T) relative =", round(rel_DCR_S, 3))

# CCR：SがTとHのどちらに近いか（Tの方が近い割合）
def ccr(SX, TX, HX, sample_S=5000):
    if len(SX) > sample_S:
        idx = np.random.choice(len(SX), size=sample_S, replace=False)
        SX = SX[idx]
    cnt_T_closer = 0
    for s in SX:
        dT = np.abs(TX - s).sum(axis=1).min()
        dH = np.abs(HX - s).sum(axis=1).min()
        cnt_T_closer += int(dT < dH)
    return cnt_T_closer / len(SX)

CCR = ccr(X_S, X_T, X_H)
print("CCR(S closer to T) =", round(CCR, 3))
```
